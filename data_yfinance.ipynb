## Extraer Precios Históricos de Acciones desde Yahoo! Finance
Varias maneras para obtener precios de acciones históricos desde Yahoo! Finance con Python utilizando el paquete yfinance (se asume que ya está instalado).

```python
# importando librerías base
import pandas as pd
import yfinance as yf
```

#### 1. yf.Ticker + .history()

```python
ticker = yf.Ticker('SBUX')
sbux = ticker.history(period='max')
sbux
```

#### 2. yf.download()

```python
msft = yf.download('MSFT', period='10y')
msft
```

#### 3. datareader + get_data_yahoo()

```python
from pandas_datareader import data as pdr
yf.pdr_override()
```

```python
# obteniendo data diaria de la empresa TexasRoad House
pdr.get_data_yahoo('TXRH', start='2012-1-1', end='2022-12-31') # esto se puede guardar en una variable
```

```python
# para obtener data semanal
pdr.get_data_yahoo('TXRH', start='2012-1-1', end='2022-12-31', interval='1wk') 
```

```python
# para obtener data mensual
pdr.get_data_yahoo('TXRH', start='2012-1-1', end='2022-12-31', interval='1mo') 
```

```python
# obteniendo precio de cierre de varias acciones
stocks = ['AAPL', 'GOOGL', 'V', 'BAC', 'T']
price_close = pdr.get_data_yahoo(stocks, start, end)['Close']
price_close
```
